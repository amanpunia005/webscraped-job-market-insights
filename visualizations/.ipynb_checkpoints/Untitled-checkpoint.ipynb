```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Aman Puniya\WebScraped_JobMarket_Insights\data\cleaned_jobs_data.csv")
df.head()
```

```python
df.tail()
```

```python
total_companies = df['Company'].str.split(',').explode().str.title()
total_companies
```

```python
count_companies = total_companies.value_counts()
top_companies = count_companies.head()
top_companies
```

```python
df1 = df[df['Experience Categories'] == 'Freshers'] 
df2 = df1['Company'].value_counts().head()
df2
```

```python
df3 = df[df['Experience Categories'] == 'Entry Level'] 
df4 = df3['Company'].value_counts().head()
df4
```

```python
df5 = df[df['Experience Categories'] == 'Mid Level'] 
df6 = df5['Company'].value_counts().head()
df6
```

```python
df7 = df[df['Experience Categories'] == 'Senior Level'] 
df8 = df7['Company'].value_counts().head()
df8
```

```python
fig, axs = plt.subplots(3,2,figsize = (25,12))

axs[0,0].barh(top_companies.index,top_companies.values,height=0.4)
axs[0,0].set_xlabel('No. of Jobs',fontsize = 15,fontweight = 'bold')
axs[0,0].set_ylabel('Companies',fontsize = 15,fontweight = 'bold')
axs[0,0].grid(axis = 'x',linestyle = '--',alpha = 0.6)
axs[0,0].invert_yaxis()
axs[0,0].set_title('Top 5 Hiring Companies',fontsize = 18,fontweight = 'bold')
axs[0,0].tick_params(axis = 'x',labelsize = 12,labelcolor = 'black')
axs[0,0].tick_params(axis = 'y',labelsize = 12,labelcolor = 'black')


axs[0,1].barh(df2.index,df2.values,height=0.4)
axs[0,1].set_xlabel('No. of Jobs',fontsize = 15,fontweight = 'bold')
axs[0,1].set_ylabel('Companies',fontsize = 15,fontweight = 'bold')
axs[0,1].grid(axis = 'x',linestyle = '--',alpha = 0.6)
axs[0,1].invert_yaxis()
axs[0,1].set_title('Top 5 Fresher Hiring Companies',fontsize = 18,fontweight = 'bold')
axs[0,1].tick_params(axis = 'x',labelsize = 12,labelcolor = 'black')
axs[0,1].tick_params(axis = 'y',labelsize = 12,labelcolor = 'black')
axs[0,1].set_xlim(0,10)


axs[1,0].barh(df4.index,df4.values,height=0.4)
axs[1,0].set_xlabel('No. of Jobs',fontsize = 15,fontweight = 'bold')
axs[1,0].set_ylabel('Companies',fontsize = 15,fontweight = 'bold')
axs[1,0].grid(axis = 'x',linestyle = '--',alpha = 0.6)
axs[1,0].invert_yaxis()
axs[1,0].set_title('Top 5 Entry Level Hiring Companies',fontsize = 18,fontweight = 'bold')
axs[1,0].tick_params(axis = 'x',labelsize = 12,labelcolor = 'black')
axs[1,0].tick_params(axis = 'y',labelsize = 12,labelcolor = 'black')

axs[1,1].barh(df6.index,df6.values,height=0.4)
axs[1,1].set_xlabel('No. of Jobs',fontsize = 15,fontweight = 'bold')
axs[1,1].set_ylabel('Companies',fontsize = 15,fontweight = 'bold')
axs[1,1].grid(axis = 'x',linestyle = '--',alpha = 0.6)
axs[1,1].invert_yaxis()
axs[1,1].set_title('Top 5 Mid Level Hiring Companies',fontsize = 18,fontweight = 'bold')
axs[1,1].tick_params(axis = 'x',labelsize = 12,labelcolor = 'black')
axs[1,1].tick_params(axis = 'y',labelsize = 12,labelcolor = 'black')

axs[2,0].barh(df8.index,df8.values,height=0.4)
axs[2,0].set_xlabel('No. of Jobs',fontsize = 15,fontweight = 'bold')
axs[2,0].set_ylabel('Companies',fontsize = 15,fontweight = 'bold')
axs[2,0].grid(axis = 'x',linestyle = '--',alpha = 0.6)
axs[2,0].invert_yaxis()
axs[2,0].set_title('Top 5 Senior Level Hiring Companies',fontsize = 18,fontweight = 'bold')
axs[2,0].tick_params(axis = 'x',labelsize = 12,labelcolor = 'black')
axs[2,0].tick_params(axis = 'y',labelsize = 12,labelcolor = 'black')


axs[2,1].axis('off')
plt.tight_layout()
```
