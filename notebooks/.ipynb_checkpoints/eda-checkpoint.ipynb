```python
df3 = df2.dropna()
```

```python
df4 = df3.drop(columns = ['Salary'])
```

```python
def new_column(exp):
    x = exp.split(" ")[0]
    try:
        if '-' in x:
            y = int(x.split('-')[0])
        else:
            y = int(x[0])
        if y == 0:
            return '0 Yrs'
        elif y <= 2:
            return '0-2 Yrs'
        elif y <= 4:
            return '2-4 Yrs'
        elif y <= 6:
            return '4-6 Yrs'
        elif y <= 8:
            return '6-8 Yrs'
        else:
            return '8+ Yrs'
    except Exception as e:
        print(f"Error : {e}")
```

```python
df4['new_experience'] = df4['Experience'].apply(new_column)
df4
```

```python
df5 = df4.drop(columns = ['Experience'])
df5
```

```python
df5.rename(columns = {'new_experience':'Experience'},inplace = True)
```

```python
def get_experience_level(exp):
    x = exp.split(" ")[0]
    try:
        if "-" in x:
            y = int(x.split("-")[0])
        else:
            y = int(x[0])
        if y == 0:
            return 'Freshers'
        elif y <= 2:
            return 'Entry Level'
        elif y <= 4 or y <= 6:
            return 'Mid Level'
        elif y <= 8:
            return 'Senior Level'
        else:
            return 'Principal'
    except Exception as e:
        print(f"Error : {e}")
        
```

```python
df5['Experience Categories'] = df5['Experience'].apply(get_experience_level)
df5
```

```python
cols = list(df5.columns)
cols.insert(1,cols.pop(cols.index('Experience')))
cols.insert(2,cols.pop(cols.index('Experience Categories')))
df5 = df5[cols]
df5
```

```python
df6 = df5.sort_values("Experience")
df6
```

```python
df7 =  df6.reset_index(drop = True)
df7
```
