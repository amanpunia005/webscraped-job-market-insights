```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Aman Puniya\WebScraped_JobMarket_Insights\data\cleaned_jobs_data.csv")
df.head()
```

```python
df2 = df['Job Title'].value_counts().sort_values(ascending=False)
df2
```

### Visualization: Top Job Titles (Log Scale)

- **Source**: Cleaned job title column from scraped listings  
- **Tool**: Seaborn barplot + Matplotlib (log scale customization)  
- **Insight**: 
  - **"Data Analyst"** dominates job postings by a wide margin.
  - Other titles like **Data Engineer**, **Data Scientist**, and **Business Analyst** have far fewer postings.
  - **Log scale** was applied to clearly visualize disparities across job roles.


```python
plt.figure(figsize=(12,5))
sns.barplot(x=df2.index, y=df2.values,color = 'steelblue',edgecolor = 'black',linewidth = 0.6)
plt.yscale('log')  # <- This compresses large values
plt.title("Top Job Titles (Log Scale)", fontsize=14, fontweight='bold')
plt.ylabel("Number of Postings (Log Scale)",fontsize = 12,fontweight = 'bold')
plt.xlabel("Job Title",fontsize = 12,fontweight = 'bold')
plt.xticks(rotation=15)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
for i,v in enumerate(df2.values):
    plt.text(i,v+3,str(v), ha = 'center')
sns.despine()
plt.tight_layout()
```
