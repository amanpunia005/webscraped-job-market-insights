```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Aman Puniya\WebScraped_JobMarket_Insights\data\cleaned_jobs_data.csv")
```

```python
all_location = df['Location'].str.split(',').explode().str.strip().str.lower()
all_location
```

```python
location_mapping = {
    'bengaluru':'benguluru',
    'bengaluru(hsr layout)':'bengaluru',
    'bangalore rural':'bangaluru',
    'bengaluru(sector 2 hsr layout)':'bengaluru',
    'bangaluru':'bengaluru',
    'hyderabad':'hyderabad',
    'new hyderabad':'hyderabad',
    'hyderabad(kokapet)':'hyderabad',
    'new delhi':'delhi',
    'delhi / ncr':'delhi',
    'new delhi(ghitorni)':'delhi',
    'mumbai':'mumbai',
    'mumbai (all areas)':'mumbai',
    'navi mumbai':'mumbai',
    'mumbai suburban':'mumbai',
    'mumbai (all areas)(prabhadevi)':'mumbai',
    'mumbai(lower parel)':'mumbai'
}
```

```python
all_location = all_location.replace(location_mapping)
location_count = all_location.value_counts()
top_location = location_count.head(10)
top_location
```

### Visualization: Top Job Locations for Data Analyst Roles
- **Source**: Cleaned job data from scraped listings
- **Tool**: Matplotlib('bar')
- **Insight**: Cities like *bangaluru*,*hyderabad*,*mumbai* dominant the hiring market for data analyst roles.

```python
plt.figure(figsize = (12,5),facecolor='white')
width = 0.70
colors = '#2e4cc9'
bars = plt.bar(top_location.index,top_location.values,alpha = 1,width = width,color = colors)
plt.xlabel("Cities",fontsize = 12,fontweight = 'bold')
plt.ylabel("Number of Job Postings",fontsize = 12,fontweight='bold')
plt.ylim(0,320)
plt.title("Top Cities for Data Analyst",fontsize=18,fontweight='bold')
plt.grid(axis = 'y',linestyle = ':',alpha = 0.5)
plt.tight_layout()
plt.xticks(rotation=45)
for i,v in enumerate(top_location.values):
    plt.text(i,v + 5, str(v),ha='center')
plt.savefig('top_job_location_vertical.png',dpi=300,bbox_inches='tight')
plt.show()
```
