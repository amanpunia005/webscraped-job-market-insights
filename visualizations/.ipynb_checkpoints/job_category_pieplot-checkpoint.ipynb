```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Aman Puniya\WebScraped_JobMarket_Insights\data\cleaned_jobs_data.csv")
df['Experience Categories']
```

```python
experience = df['Experience Categories'].str.split(',').explode().str.strip().str.title()
experience
```

```python
category_count = experience.value_counts()
category_count
```

### Visualization: Job Distribution by Role Seniority
- **Source:** Cleaned job data from scraped listings
- **Tool:** Matplotlib (pie chart), Seaborne (styling/colors)
- **Insight:** Majority of job roles target **Freshers**,followed by **Entry level** and **mid Level** positions.Very few listings are for **Senior Level** roles.

```python
plt.figure(figsize=(7,7),facecolor="white")
labels = (category_count.index)
colors = sns.color_palette('muted')[0:5]
plt.pie(category_count,labels = labels,textprops={'fontsize':12,'color':'black'},colors=colors,wedgeprops={'edgecolor': 'black', 'linewidth': 1},autopct='%1.1f%%')
plt.title('% of Job per Category',fontsize = 26,pad=10,fontweight = 'bold')
sns.set_style("darkgrid")
plt.legend(category_count.index,loc='best',bbox_to_anchor=(1,0.9))
plt.axis('equal')
plt.tight_layout()
plt.savefig('job_category_pie.png',dpi=300,bbox_inches='tight')
plt.show(); 
```
