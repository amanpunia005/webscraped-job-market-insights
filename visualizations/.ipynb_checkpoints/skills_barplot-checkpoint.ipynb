```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Aman Puniya\WebScraped_JobMarket_Insights\data\cleaned_jobs_data.csv")
```

```python
df.head()
```

```python
all_skills = df['Skills'].str.split(',').explode().str.strip().str.lower()
skill_mapping = {
    'data analyst': 'data analysis',
    'data analytics': 'data analysis',
    'analytical': 'data analysis',
    'data': 'data analysis',
    'powerbi': 'power bi',
    'Power Bi' : 'power bi',
    'excel.': 'excel',
    'Excel' : 'excel',
    'python.': 'python',
    'Python':'python',
    'sql.': 'sql',
    'SQL' : 'sql',
    'machine learning.': 'machine learning',
    'ms excel': 'excel',
    'Sql': 'sql',
    'python programming': 'python',
    'data science': 'data analysis',
    'visualisation': 'data visualization',
    'visualizations': 'data visualization'
}
all_skills = all_skills.replace(skill_mapping)
```

```python
skills_count = all_skills.value_counts()
skills = skills_count.head(10)
```

## Visualization: Top In-Demand Skills for Data Analyst Roles
- **Source**: Cleaned job data from scraped listings
- **Tool**: Matplotlib(barh)
- **Insight**: "Data Analysis","SQL"and "Python" are top-requested skills.

```python
plt.figure(figsize=(12,5),facecolor='white') 
colors = '#7629d9'
bars = plt.barh(skills.index, skills.values,alpha=0.9,color = colors)
plt.xlabel("Number of Job Postings",fontsize = 14,fontweight='bold')
plt.ylabel("Skills",fontsize = 14,fontweight='bold')
plt.title("Top In-Demand Skills in Data Analyst Job Market",fontsize = 18,fontweight='bold')
plt.xlim(0,1500)
plt.xticks(color='black')
plt.yticks(color='black')
plt.tight_layout()
plt.gca().invert_yaxis()
plt.grid(axis = 'x',linestyle = '--',alpha = 0.5)
for i,v in enumerate(skills.values):
    plt.text(v+30,i,str(v),ha='center')
plt.savefig('top_skills_horizontal.png',dpi=300,bbox_inches='tight')
plt.show()
```
